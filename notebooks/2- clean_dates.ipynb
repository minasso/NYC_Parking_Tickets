```python
import pandas as pd
```

```python
pd.set_option('display.min_rows', 500)  # <-add this!
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv('Parking_2016-2018_20241102.csv')
```

```python
df.columns = df.columns.map(lambda x:x.lower())
```

```python
df.columns = ['state', 'license', 'date', 'time', 'violation',
       'precinct', 'county', 'agency', 'fine']
```

```python
df['state'] = df.state.astype('category')
df['license'] = df['license'].astype('category')
df['violation'] = df['violation'].astype('category')
df['precinct'] = df['precinct'].astype('category')
df['county'] = df['county'].astype('category')
df['agency'] = df['agency'].astype('category')
```

```python
from datetime import datetime

def validate(date_text):
    try:
        if date_text != datetime.strptime(date_text, "%m/%d/%Y").strftime('%m/%d/%Y'):
            raise ValueError
        return True
    except ValueError:
        return False
```

```python
df.date = df.date.str.strip()
```

```python
df.date = df.date.str.replace('T00:00:00.000','')
```

```python
df.date = df.date.str.replace('(\d{4})-(\d{2})-(\d{2})',r'\2/\3/\1')
```

```python
df.date
```

Working on time now

```python
df.time
```

```python
df.time = df.time.str.replace('(\d{2}):(\d{2})([AP])',r'\1:\2 \3M')
```

```python
df.time
```

```python
df
```

```python
df['datetime'] = df.date + " " + df.time
```

```python
df
```

```python
df['dt'] = pd.to_datetime(df['datetime'], errors='coerce')
```

```python
df
```

```python
df.dt.isnull().sum()
```

```python
df = df.dropna(subset='dt')
```

```python
df.drop('datetime',axis=1,inplace=True)
```

```python
df.to_csv('parking_clean_datetimes.csv',index=False)
```

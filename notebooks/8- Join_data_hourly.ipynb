```python
import pandas as pd
```

```python
f = pd.read_pickle('../data/interim/parking_clean_dtypes.p')
```

```python
f
```

```python
f['hour'] = f.dt.dt.hour
```

```python
new = f.groupby(['date','hour']).count()
```

```python
new
```

```python
n = new.reset_index()
```

```python
n
```

```python
n = n.loc[n['date'] != '19/10/2018']
```

```python
n
```

```python
n['date'] = pd.to_datetime(n.date)
```

```python
n['time'] = pd.to_datetime(n['hour'],format="%H").dt.time
```

```python
n
```

```python
n['date'] = n['date'].astype(str)
n['time'] = n['time'].astype(str)
n['datetime'] = pd.to_datetime(n['date'] + ' ' + n['time'])
```

```python
n
```

```python
n = n.set_index('datetime',drop=True)
```

```python
n
```

```python
n = n[['state']]
```

```python
n.columns = ['ticket_count']
```

```python
n
```

```python
df = pd.read_pickle('../data/interim/temp_clean.p')
```

```python
df
```

```python
fr = df.join(n,how='inner')
```

```python
fr
```

```python
pd.to_pickle(fr,'../data/interim/joined_data_hourly.p')
```

```python
import pandas as pd
```

```python
df = pd.read_csv('../data/external/NYC_Weather_2016_2022.csv')
```

```python
df[['time', 'temperature_2m (°C)', 'precipitation (mm)', 'rain (mm)',
       'cloudcover (%)', 'cloudcover_low (%)', 'cloudcover_mid (%)',
       'cloudcover_high (%)', 'windspeed_10m (km/h)', 'winddirection_10m (°)']]
```

```python
df.isna().sum()
```

Amount of missing data is pretty insignificant so we will drop it. Unsure if it is missing completely at random or not, so this may have an effect on causal analysis.

```python
df = df.dropna()
```

```python
df
```

```python
df.info()
```

```python
df['time'] = pd.to_datetime(df['time'])
```

```python
df.index = df['time']
```

```python
df = df.drop('time',axis=1)
```

```python
df['temp'] = df['temperature_2m (°C)']*9/5 + 32
```

```python
df = df.drop('temperature_2m (°C)',axis=1)
```
